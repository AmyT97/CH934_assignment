```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("time_vs_cpu_data.csv")
```

```python
df.head()
```

```python
ant = df[df["Molecule "] == "ANTCEN"]
pen = df[df["Molecule "] == "PENCEN"].reset_index(drop=True)
tet = df[df["Molecule "] == "TETCEN"].reset_index(drop=True)
```

```python
plt.style.use("dark_background")

plt.plot(ant["Core"], ant["Time"], marker="o", label="ANTCEN", color="orange")
plt.plot(pen["Core"], pen["Time"], marker="o", label="PENCEN", color="green")
plt.plot(tet["Core"], tet["Time"], marker="o", label="TETCEN", color="blue")

plt.xlabel("CPU Core Count")
plt.ylabel("Execution Time (s)")
plt.title("Execution Time vs CPU Cores")

plt.legend()
plt.grid(True, linestyle="--", alpha=0.3)

plt.show()
```

```python
#speed up is time of the baseline (2 core run) divided by the time of another run (n)
# speedup = ant["Time"][0]/ant["Time"][n]

ant_speedup = []

for n in range (1, len(ant)):
    s = ant["Time"][0] / ant["Time"][n]
    ant_speedup.append(f"{s:.2f}")
    
print("ANTCEN Speedup Relative to 2 Core Run:")
print("4 Core:", ant_speedup[0])
print("8 Core:", ant_speedup[1])
print("16 Core:", ant_speedup[2])
```

```python
#speed up is time of the baseline (2 core run) divided by the time of another run (n)
# speedup = ant["Time"][0]/ant["Time"][n]

pen_speedup = []

for n in range (1, len(pen)):
    s = pen["Time"][0] / pen["Time"][n]
    pen_speedup.append(f"{s:.2f}")
    
print("PENCEN Speedup Relative to 2 Core Run:")
print("4 Core:", pen_speedup[0])
print("8 Core:", pen_speedup[1])
print("16 Core:", pen_speedup[2])
```

```python
#speed up is time of the baseline (2 core run) divided by the time of another run (n)
# speedup = ant["Time"][0]/ant["Time"][n]

tet_speedup = []

for n in range (1, len(tet)):
    s = tet["Time"][0] / tet["Time"][n]
    tet_speedup.append(f"{s:.2f}")
    
print("TETCEN Speedup Relative to 2 Core Run:")
print("4 Core:", tet_speedup[0])
print("8 Core:", tet_speedup[1])
print("16 Core:", tet_speedup[2])
```

```python
#parallel efficiency is the speedup divided by the number of cores
cores = [4,8,16]

ant_efficiency = []

for n in range (len(ant_speedup)):
    e = float(ant_speedup[n]) / cores[n]
    ant_efficiency.append(f"{e:.2f}")

print(ant_efficiency)
```

```python
#parallel efficiency is the speedup divided by the number of cores
cores = [4,8,16]

pen_efficiency = []

for n in range (len(pen_speedup)):
    e = float(pen_speedup[n]) / cores[n]
    pen_efficiency.append(f"{e:.2f}")

print(pen_efficiency)
```
